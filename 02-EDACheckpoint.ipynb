## Research Question

The main question we aim to answer is: To what extent does a PhD provide a statistically significant decrease in AI-driven automation risk, specifically when compared to 10 years of professional experience? And how exactly does this "shielding effect" vary between technical and service sectors? Is it built on a house of cards, or can it be used to gauge job security reliably?

**Dependent Variable:** Automation Probability 2030 / AI Exposure Index  
**Independent Variables:** Education Level (PhD vs. less than PhD), Years of Experience  
**Moderating Variable:** Technical Job Sector, Service Job Sector  
**Control Variables:** Baseline Skill Profiles, Professional Networks/Connections, Existing Industry-Specific Automation Levels


## Background and Prior Work

The rapid integration of Generative AI into the global workforce has sparked an intense debate about which demographics are most vulnerable to job displacement. Unlike earlier waves of automation that primarily replaced manual labor, today’s AI systems generate text, code, and images, allowing them to perform cognitive tasks once associated with college-educated professionals. As adoption has accelerated since 2023, students and workers are questioning whether advanced degrees still provide protection or whether years of professional experience now offer a stronger defense. With 2030 projections approaching, it is no longer clear that a Master’s degree or even a PhD guarantees career stability, nor whether a decade of professional experience offers a stronger buffer.

This project specifically investigates whether a PhD provides a statistically significant reduction in AI-driven automation probability when compared to 10 years of professional experience. It also evaluates whether this potential shielding effect varies across technical and service sectors. By structuring the analysis around automation probability and AI exposure indices, we move beyond general discussion and instead test whether formal education or accumulated experience serves as a more reliable predictor of job security.

The theoretical foundation for this analysis comes from the task-based framework developed by Daron Acemoglu and Pascual Restrepo. Their framework argues that the impact of automation depends on the specific tasks within a job rather than the job title itself. Automation has a comparative advantage in tasks where machines can process information more efficiently, which explains why certain cognitive tasks are now highly exposed to AI. At the same time, technological change introduces new, complex tasks that favor non-routine cognitive skills, potentially creating a shielding effect for highly trained individuals. However, if AI primarily automates routine cognitive tasks that are heavily represented in technical professions, then advanced degrees may not provide the protection traditionally assumed. In service sectors, experience may matter more because these roles depend on social intelligence, tacit knowledge, and situational awareness, areas where AI remains comparatively limited.

Empirical evidence reflects this tension. Tyna Eloundou et al. (2023) estimate that roughly 80 percent of the U.S. workforce could see at least 10 percent of their tasks affected by large language models, with higher exposure among workers holding advanced degrees due to the automaticity of writing and programming tasks. This challenges the assumption that more formal education linearly reduces risk. Conversely, research by Erik Brynjolfsson, Danielle Li, and Lindsey Raymond (2023) highlights complementarity effects, showing that AI can significantly boost productivity and act as a collaborator for experienced professionals rather than a substitute. The OECD’s 2023 Employment Outlook similarly identifies sectoral divergence, where AI substitutes for routine technical tasks but struggles to replace high-touch service work.

Although projections about 2030 remain inherently uncertain, this project moves beyond speculation by applying task-based modeling with O*NET data and Kaggle’s AI Impact on Jobs 2030 dataset to statistically test whether a PhD or ten years of experience provides a stronger buffer against automation risk, and whether that buffering effect depends on sector.

## Hypothesis

We hypothesize that a PhD will provide a greater reduction in automation risk than 10 years of professional experience within technical sectors. However, we expect the opposite pattern in service sectors, where accumulated experience will be more protective than formal education. 

This hypothesis is grounded in the assumption that technical sectors reward specialized, high-level theoretical knowledge that may complement AI systems, while service sectors depend more heavily on practical judgment, tacit knowledge, and interpersonal skills that develop through experience and remain more difficult for AI to replicate. Therefore, we expect a sector-dependent shielding effect in which the relative value of education versus experience varies systematically across occupational contexts.

### Data Overview

#### Dataset #1

- **Dataset Name:** AI Impact on Jobs 2030  
- **Link to the dataset:** https://www.kaggle.com/datasets/khushikyad001/ai-impact-on-jobs-2030  
- **Number of observations:** 3000  
- **Number of variables:** 18  

**Description of the variables most relevant to this project:**

**Education_Level**  
Categorical variable showing the highest formal education required for the role. This represents the formal credential side of the model. I isolate the PhD category and compare its automation risk to other education levels to test whether the degree provides a measurable shielding effect.

**Years_Experience**  
Numerical variable representing the average years of experience tied to the job. This represents the practical experience side of the model. I use it to group higher-experience roles and compare them directly to PhD-level roles.

**Automation_Probability_2030**  
Continuous variable between 0 and 1 estimating the likelihood of automation by 2030. This is the primary dependent variable. All statistical comparisons center on this measure of risk.

**Job_Title**  
Specific occupation name. Since the dataset does not include an explicit sector variable, I use job titles to manually classify occupations into technical and service sectors for moderation analysis.

**AI_Exposure_Index**  
Score measuring how integrated AI is into daily job tasks. High exposure does not automatically imply high automation risk. This variable helps distinguish augmentation from replacement.

**Skill_1, Skill_2, …, Skill_10**  
Numerical skill scores representing underlying capability profiles. These act as control variables, allowing analysis of whether protection comes from the degree itself or from the associated skill composition.

**Risk_Category**  
Categorical label (Low, Medium, High) derived from automation probability. This is useful for validating internal consistency of the dataset.

**Shortcomings of the Dataset:**

- The data is synthetic, meaning any observed shielding effect may reflect built-in modeling assumptions rather than real labor market outcomes.
- Skills are labeled generically, limiting the ability to identify which specific human capabilities drive protection from automation.
- There is no predefined sector variable. Technical versus service classifications must be created manually, introducing subjectivity into the moderation analysis.
