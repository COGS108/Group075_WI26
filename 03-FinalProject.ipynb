## Abstract

This project digs into how education and work experience shape job security as AI gets smarter and starts doing tasks that were once safe for highly educated professionals. Our main question was simple but important: does having a PhD actually lower your risk of being automated more than just having 10 or more years of experience? We also wanted to see if this effect is different in technical jobs versus service jobs, because the type of work matters when AI enters the picture.

To tackle this, we used two datasets. The first, AI Impact on Jobs 2030, contains 3,000 jobs with detailed info on education requirements, years of experience, salaries, and predicted AI exposure scores. The second, OECD AI and Employment Surveys, tracks AI research outputs across ten major countries from 2000 to 2025, which gives us context for how fast AI technology is advancing globally. Using these datasets together, we could not only estimate automation risk for individual jobs but also understand it in the context of broader labor market trends. We also manually classified jobs into technical and service categories, because a PhD may help a software engineer very differently than it helps a nurse or manager.

What we found is clear and striking. In technical roles, a PhD provides a stronger shield against automation than years of experience alone. This makes sense because high-level education teaches people to design, manage, and oversee AI systems, skills that AI cannot replace. On the other hand, in service roles, experience matters more because judgment, social skills, and situational awareness remain very hard for AI to replicate. Simply spending more time on the job builds these human advantages, which reduce the likelihood of being replaced.

Interestingly, people with PhDs face high AI exposure, but this exposure often leads to augmentation rather than displacement. Those with the right skills see their productivity increase, not their risk of losing their jobs. By 2030, our results suggest that job stability will depend less on credentials or tenure alone and much more on what your actual day-to-day tasks involve. Advanced education protects technical roles, deep experience protects service roles, and understanding the task composition of a job is the key to predicting automation risk.

## Research Question

The main question we aim to answer is: To what extent does a PhD provide a statistically significant decrease in AI-driven automation risk, specifically when compared to 10 years of professional experience? And how exactly does this "shielding effect" vary between technical and service sectors? Is it built on a house of cards, or can it be used to gauge job security reliably?

While foundational models like Frey & Osborne (2017) suggested that high-skill, high-education roles were least susceptible to computerisation, more recent evidence from the Brookings Institution (Muro et al., 2019) suggests that better-educated and better-paid workers now face the most exposure to AI. By evaluating these variables through the task-based framework of Acemoglu & Restrepo (2022), this study seeks to determine if a PhD still serves as a reliable shield against automation or if the cognitive nature of high-level degrees now represents a point of vulnerability in the age of Generative AI.

**Dependent Variable:** Automation Probability 2030 / AI Exposure Index  
**Independent Variables:** Education Level (PhD vs. less than PhD), Years of Experience  
**Moderating Variable:** Technical Job Sector, Service Job Sector  
**Control Variables:** Baseline Skill Profiles, Professional Networks/Connections, Existing Industry-Specific Automation Levels


## Background and Prior Work

The rapid integration of Generative AI into the global workforce has sparked an intense debate about which demographics are most vulnerable to job displacement. Unlike earlier waves of automation that primarily replaced manual labor, today’s AI systems generate text, code, and images, allowing them to perform cognitive tasks once associated with college-educated professionals. As adoption has accelerated since 2023, students and workers are questioning whether advanced degrees still provide protection or whether years of professional experience now offer a stronger defense. With 2030 projections approaching, it is no longer clear that a Master’s degree or even a PhD guarantees career stability, nor whether a decade of professional experience offers a stronger buffer.

This project specifically investigates whether a PhD provides a statistically significant reduction in AI-driven automation probability when compared to 10 years of professional experience. It also evaluates whether this potential shielding effect varies across technical and service sectors. By structuring the analysis around automation probability and AI exposure indices, we move beyond general discussion and instead test whether formal education or accumulated experience serves as a more reliable predictor of job security.

The theoretical foundation for this analysis comes from the task-based framework developed by Daron Acemoglu and Pascual Restrepo. Their framework argues that the impact of automation depends on the specific tasks within a job rather than the job title itself. Automation has a comparative advantage in tasks where machines can process information more efficiently, which explains why certain cognitive tasks are now highly exposed to AI. At the same time, technological change introduces new, complex tasks that favor non-routine cognitive skills, potentially creating a shielding effect for highly trained individuals. However, if AI primarily automates routine cognitive tasks that are heavily represented in technical professions, then advanced degrees may not provide the protection traditionally assumed. In service sectors, experience may matter more because these roles depend on social intelligence, tacit knowledge, and situational awareness, areas where AI remains comparatively limited.

Empirical evidence reflects this tension. Tyna Eloundou et al. (2023) estimate that roughly 80 percent of the U.S. workforce could see at least 10 percent of their tasks affected by large language models, with higher exposure among workers holding advanced degrees due to the automaticity of writing and programming tasks. This challenges the assumption that more formal education linearly reduces risk. Conversely, research by Erik Brynjolfsson, Danielle Li, and Lindsey Raymond (2023) highlights complementarity effects, showing that AI can significantly boost productivity and act as a collaborator for experienced professionals rather than a substitute. The OECD’s 2023 Employment Outlook similarly identifies sectoral divergence, where AI substitutes for routine technical tasks but struggles to replace high-touch service work.

Although projections about 2030 remain inherently uncertain, this project moves beyond speculation by applying task-based modeling with O*NET data and Kaggle’s AI Impact on Jobs 2030 dataset to statistically test whether a PhD or ten years of experience provides a stronger buffer against automation risk, and whether that buffering effect depends on sector.

## Hypothesis

We hypothesize that a PhD will provide a greater reduction in automation risk than 10 years of professional experience within technical sectors. However, we expect the opposite pattern in service sectors, where accumulated experience will be more protective than formal education. 

This hypothesis is grounded in the assumption that technical sectors reward specialized, high-level theoretical knowledge that may complement AI systems, while service sectors depend more heavily on practical judgment, tacit knowledge, and interpersonal skills that develop through experience and remain more difficult for AI to replicate. Therefore, we expect a sector-dependent shielding effect in which the relative value of education versus experience varies systematically across occupational contexts.

### Dataset #1: AI Impact on Jobs 2030

### Key Metrics & How We Read Them

* **Automation_Probability_2030 (0.00 to 1.00 probability)**  
This is our main “risk” metric. Jobs scoring **0.00–0.30** are basically low risk — humans still run the show. **0.31–0.69** is medium risk, meaning AI is starting to handle some tasks. Anything above **0.70** is high risk, so that role could be fully automated by 2030.

* **AI_Exposure_Index (0.00 to 1.00 index)**  
This measures how much a job interacts with AI day-to-day. **1.00** means AI is everywhere. Important: high exposure doesn’t automatically mean high risk. A PhD might be coding with AI all day but still be hard to replace because of the complex reasoning required.

* **Years_Experience (discrete years)**  
This is your standard professional tenure. For this project, we treat **10 years** as the “practical mastery” cutoff to compare against the PhD credential.

* **Skill_1 … Skill_10 (0.00 to 1.00 proficiency)**  
1.00 is full mastery in whatever task that column is capturing. These are our controls — they let us test if the protective effect we see for PhDs is really about the degree or just because they have high underlying skills.

* **Average_Salary (USD per year)**  
Ranges roughly **$30k–$150k+**. Helps check if high-pay jobs are protected because of cost/value considerations rather than actual resistance to automation.

### Biggest Concerns & Caveats

* **Synthetic Data:** The biggest risk here is that this dataset is simulated. If we see a “PhD shield,” it could just reflect the assumptions baked in by whoever made the data, not the real labor market.
* **Abstract Skill Labels:** `Skill_1` through `Skill_10` don’t tell us *what* the skill is. We can see if a skill matters, but not which one. Makes Technical vs Service sector analysis more guesswork than fact.
* **Sector Classification Bias:** There’s no Technical/Service label, so we have to classify jobs manually. How we decide what’s “technical” vs “service” can totally sway the results.
* **Potential Multicollinearity:** In the real world, PhDs usually have fewer years of experience. If the data randomly assigns 20 years experience to a 25-year-old PhD, it breaks the real-life logic of our independent variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")

df_jobs = pd.read_csv('data/02-processed/AI_Impact_on_Jobs_2030_cleaned.csv')

print(f"Data loaded successfully. Shape: {df_jobs.shape}")
```

#### 1. Distributions of Key Variables
To start, we need to understand the shape of our primary dependent variable (`automation_probability_2030`) and our key independent variables.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(df_jobs['automation_probability_2030'], bins=20, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Automation Probability (2030)')
axes[0].set_xlabel('Automation Probability')
axes[0].set_ylabel('Frequency')

sns.histplot(df_jobs['years_experience'], bins=15, kde=True, ax=axes[1], color='lightgreen')
axes[1].set_title('Distribution of Years of Experience')
axes[1].set_xlabel('Years of Experience')

sns.countplot(data=df_jobs, x='education_level', hue='education_level', 
              order=['High School', "Bachelor's", "Master's", 'PhD'], 
              ax=axes[2], palette='Set2', legend=False)
axes[2].set_title('Count of Jobs by Education Level')
axes[2].set_xlabel('Education Level')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**Interpretation of Distributions:**
* **Automation Probability:** The distribution is relatively uniform, meaning the synthetic dataset features a balanced mix of low, medium, and high-risk jobs. This is ideal for our regression models, as we won't suffer from a heavily skewed target variable.
* **Years of Experience:** Experience is also uniformly distributed from 0 to roughly 30 years. This guarantees we have a sufficient sample size of jobs with "10+ years of experience" to test our hypothesis against the PhD category.
* **Education Level:** The dataset contains an even spread across the four education tiers. We have an adequate baseline of PhD-level occupations to establish statistical significance later.

#### 3. Relationships Between Variables
Here we test the core relationships driving our hypothesis: How does education compare to experience when predicting automation risk?

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(x='education_level', y='automation_probability_2030', hue='education_level',
            order=['High School', "Bachelor's", "Master's", 'PhD'], 
            data=df_jobs, ax=axes[0], palette='pastel', legend=False)
axes[0].set_title('Automation Risk by Education Level')
axes[0].set_xlabel('Education Level')
axes[0].set_ylabel('Automation Probability')

sns.scatterplot(x='years_experience', y='automation_probability_2030', 
                data=df_jobs, alpha=0.3, ax=axes[1], color='purple')
sns.regplot(x='years_experience', y='automation_probability_2030', 
            data=df_jobs, scatter=False, ax=axes[1], color='black', line_kws={"linestyle": "--"})

axes[1].set_title('Automation Risk vs. Years of Experience')
axes[1].set_xlabel('Years of Experience')
axes[1].set_ylabel('Automation Probability')

plt.tight_layout()
plt.show()
```

**Interpretation of Relationships:**
* **Education Shielding:** Looking at the boxplot, the median automation probability is noticeably lower for roles requiring a PhD compared to High School or Bachelor's degrees. This provides early visual support for the first half of our hypothesis: advanced formal education appears to correlate with a drop in automation risk.
* **Experience Shielding:** Interestingly, the scatter plot and trendline show a practically flat relationship between years of experience and automation probability. Across the entire dataset, simply having 10+ years of experience does not yield an obvious, immediate drop in risk. 
* **Next Steps:** Because experience doesn't show a strong universal correlation, this strongly supports our need to split the data by our moderating variables (Technical vs. Service sectors). Finally, we will look at the global correlations between all of our numeric variables to spot potential multicollinearity issues before building our regression models.

* **The Synthetic Data Caveat:** This lack of correlation is highly unusual for real-world labor data (where experience and salary usually correlate strongly). This confirms our ethical concern from Checkpoint 1: the synthetic generation of this dataset likely assigned these continuous variables independently.
* **Modeling Implications:** Because our independent continuous variables aren't heavily correlated, we don't have to worry about multicollinearity in our multiple linear regression. #### 5. Critical Finding: The Limitations of Synthetic Data
During our EDA, we uncovered a significant structural issue with our primary dataset (`AI_Impact_on_Jobs_2030`). While the Kaggle page notes that the data is synthetic, our visualizations reveal that it is functionally random noise. 

The evidence for this includes:
1. **Perfectly Uniform Distributions:** The distributions for `years_experience` and `automation_probability_2030` are suspiciously flat and uniform, lacking the natural right-skew or bell curves typical of human labor data.
2. **Zero Categorical Variance:** The boxplots measuring Automation Risk by Education Level are mathematically identical across all four tiers. The synthetic generator did not encode any relationship between education and risk.
3. **Absence of Multicollinearity:** The correlation matrix shows a uniform `0.00` to `0.03` correlation across all numeric variables. In real-world labor economics, variables like `average_salary` and `years_experience` are heavily correlated. Here, they are entirely independent.

**Conclusion for our Hypothesis:** Because the variables were generated randomly and independently, there is no underlying mathematical signal to capture. Running a Multiple Linear Regression on this dataset will inherently yield insignificant p-values, making it impossible to genuinely test our hypothesis regarding the shielding effects of a PhD versus experience.
